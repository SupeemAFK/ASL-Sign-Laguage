```python
import tensorflow as tf
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import cv2
import os

os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

# Load data

```python
data = tf.keras.utils.image_dataset_from_directory("data/asl_alphabet_train/asl_alphabet_train", batch_size=32, image_size=(100, 100))
batch = data.as_numpy_iterator().next()
labels = data.class_names
batch[0].shape
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
```

```python
train_data = data.take(9)
val_data = data.skip(1).take(1)
```

# Build Model

```python
import tensorflow_hub as hub
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, Activation, MaxPooling2D

base_model = tf.keras.applications.vgg16.VGG16(input_shape=(100, 100, 3), include_top=False, weights='imagenet')
base_model.trainable = False

model = Sequential()

model.add(base_model)

model.add(Flatten())

model.add(Dense(256, activation='relu'))

model.add(Dense(29, activation='softmax'))
```

```python
model.compile('adam',
              loss=tf.keras.losses.sparse_categorical_crossentropy,
              metrics=['accuracy'])
```

```python
model.summary()
```

# Traing Model

```python
history = model.fit(train_data, epochs=50, validation_data=val_data)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

```python
test_data = tf.keras.utils.image_dataset_from_directory("data/asl_alphabet_test/asl_alphabet_test", batch_size=32, image_size=(100, 100))

print("Evaluate on test data")
results = model.evaluate(test_data)
print("test loss: {loss}, test acc: {acc}".format(loss=results[0], acc=results[1]))
```

```python
# import the opencv library
import cv2
import mediapipe as mp
  
# define a video capture object
vid = cv2.VideoCapture(0)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, 800)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, 600)

#define hand detection
mp_hand = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
hands = mp_hand.Hands()

offset = 30

while(True):
    
    # Capture the video frame
    # by frame
    ret, frame = vid.read()
    frame = cv2.flip(frame, 1)
    h, w, c = frame.shape

    result = hands.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    if result.multi_hand_landmarks:
        for hand_landmark in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(
              frame,
              hand_landmark,
              mp_hand.HAND_CONNECTIONS,
              mp_drawing_styles.get_default_hand_landmarks_style(),
              mp_drawing_styles.get_default_hand_connections_style())
            
            x_max = 0
            y_max = 0
            x_min = w
            y_min = h
            for lm in hand_landmark.landmark:
                x, y = int(lm.x * w), int(lm.y * h)
                if x > x_max:
                    x_max = x + offset
                if x < x_min:
                    x_min = x - offset
                if y > y_max:
                    y_max = y + offset
                if y < y_min:
                    y_min = y - offset
            
            #Draw rectangle
            cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
            
            #Crop image frame
            crop_frame= frame[y_min: y_max, x_min: x_max]
            height, width, channels = crop_frame.shape
            if height > 0 and width > 0:
                resize = tf.image.resize(crop_frame, (100, 100))
                yhat = model.predict(np.expand_dims(resize, axis=0))
                cv2.putText(frame, labels[yhat.argmax()], (x_max, y_max), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 2)
                cv2.imshow('crop_frame', crop_frame)
                    
    # Display the resulting frame
    cv2.imshow('frame', frame)
      
    # the 'q' button is set as the
    # quitting button you may use any
    # desired button of your choiceq
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
  
# After the loop release the cap object
vid.release()
# Destroy all the windows
cv2.destroyAllWindows()
```
